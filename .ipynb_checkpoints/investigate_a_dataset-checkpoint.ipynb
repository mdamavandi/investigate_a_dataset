## Introduction

The datasets I've chosen are from Gapminder World. The first details the hourly compensation for workers throughout the world. The second details the average working hours per week of workers around the world. The last dataset is the trade balances of these countries. Each of these datasets is broken out by country and year.

There has been a lot of talk (from the Executive Branch of the US government) about how the trade balance of the United States has negatively affected the country. I wanted to explore how it may have affected the workers of this country. I also wanted to explore which countries were performing better relative to their peers in terms of wages and working hours, and worse in terms of trade balance.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from copy import deepcopy
%matplotlib inline
```

## Data Wrangling
Below I will load the data and do some plenary checks for cleanliness. I'll do some data munging to make sure the data will be tidy enough to work with.

### General Properties

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.
compensation = pd.read_csv('hourly_compensation.csv')
compensation.head()
```

```python
weekly_hours = pd.read_csv('hours_worked_per_week.csv')
weekly_hours.head()
```

```python
trade_balance = pd.read_csv('trade_balance.csv')
trade_balance.head()
```

We can see in these data sets that there are some countries that did not report statistics consistently over the years. We will need to remove these countries from consideration in our analysis.

### Data Cleaning (Replace this with more specific notes!)

We will start by unpivoting the data, so that the year dimension becomes a columnar trait. We'll also drop rows for which there is no data.

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.
def unpivot(data, valueName):
    return data.melt(id_vars=['country'], var_name='year', value_name=valueName).dropna(subset=[valueName])
```

```python
compensation = unpivot(compensation, 'hourly_rate')
weekly_hours = unpivot(weekly_hours, 'working_hours')
trade_balance = unpivot(trade_balance, 'trade_balance')
```

Combining the data will allow us to compare two traits that share a country and year. We'll also limit our combined dataset to countries that reported every five years starting in 1980 so that we can get a feel for how these traits changed over time.

```python
combined = compensation.merge(
    weekly_hours, 
    on=['country','year'], 
    how='inner'
).merge(
    trade_balance, 
    on=['country','year'], 
    how='inner'
)

combined = combined[combined['year'].isin(['1980','1985','1990','1995','2000','2005'])]
counts = combined[combined['year'].isin(['1980','1985','1990','1995','2000','2005'])].groupby('country').size()
combined = combined[combined['country'].isin(list(counts.where(counts > 5).dropna().index))]
```

## Exploratory Data Analysis

### Research Question 1: Top Performers for Workers, Worst Performer for Trade

Which of the countries are performing better in wages and working hours? Which country is performing worse in trade balance? We'll now calculate the percent change in wages & working hours, and the rate of change of the trade balance. (I explored calculating the percent change for trade balance, but because the numbers are very large, comparatively small changes made it difficult to interpret the results.) Best cases for workers would be increased wages and lowered working hours, worst case for trade would be a higher trade deficit.

```python
trade = deepcopy(combined.pivot(index='country', columns='year', values='trade_balance'))
wage = deepcopy(combined.pivot(index='country', columns='year', values='hourly_rate'))
work = deepcopy(combined.pivot(index='country', columns='year', values='working_hours'))
```

```python
i = 1
trade_delta = {}
wage_delta = {}
work_delta = {}
year_list = ['1985','1990','1995','2000','2005']
while i < len(trade.iloc[0]):
    trade_delta[year_list[i-1]] = (trade.iloc[:,i] - trade.iloc[:,i-1]) / 5 # rate of change for every 5 years
    wage_delta[year_list[i-1]] = (wage.iloc[:,i] - wage.iloc[:,i-1]) / abs(wage.iloc[:,i-1]) * 100 # % change for every 5 years
    work_delta[year_list[i-1]] = (work.iloc[:,i] - work.iloc[:,i-1]) / abs(work.iloc[:,i-1]) * 100 # % change for every 5 years
    i += 1
```

```python
print(
    'Best performer for worker wages: {} with an average {:.2f}% change'.format(
        pd.DataFrame(wage_delta).mean(axis=1).abs().idxmax(),
        pd.DataFrame(wage_delta).mean(axis=1).max()
    )
)
print(
    'Best performer for worker hours: {} with an average {:.2f}% change'.format(
        pd.DataFrame(work_delta).mean(axis=1).abs().idxmax(),
        pd.DataFrame(work_delta).mean(axis=1).min()
    )
)
print(
    'Worst performer for trade balance: {} with an average ${:,.0f} change every five years'.format(
        pd.DataFrame(trade_delta).mean(axis=1).abs().idxmax(),
        pd.DataFrame(trade_delta).mean(axis=1).min()
    ).replace('$-', '-$')
)
```

Below are some boxplots that show the Korea, Rep. and the United States relative to their peers.

```python
def boxplot_with_labels(trait, country, offset):
    plt.figure(num=None, figsize=(12,11), dpi=80)
    ax = sns.boxplot(x='year', y=trait, data=combined, whis=np.inf)
    ax = sns.swarmplot(x='year', y=trait, hue='country', data=combined)
    for line in range(0,combined[combined.country == country].shape[0]):
        ax.text(line, combined[combined.country == country][trait].iloc[line]+offset, country, horizontalalignment='center', size='medium', color='black', weight='semibold')
    l = ax.legend()
    ax.legend(loc='best', frameon=False)
```

```python
boxplot_with_labels('hourly_rate', pd.DataFrame(wage_delta).mean(axis=1).abs().idxmax(), -0.75)
```

```python
boxplot_with_labels('working_hours', pd.DataFrame(work_delta).mean(axis=1).abs().idxmax(), 0.35)
```

```python
boxplot_with_labels('trade_balance', pd.DataFrame(trade_delta).mean(axis=1).abs().idxmax(), -20000000000)
```

### Research Question 2: Trade Balance vs. Workers

What does the trade balance actually mean for workers? This is a difficult question with many possible analyses and answers. Below I'll use regression plots to examine the correlation of trade balance to wages and working hours.

```python
def correlation_plot(x_trait, y_trait):
    plt.figure(num=None, figsize=(12,11), dpi=80)
    ax = sns.regplot(x=x_trait, y=y_trait, data=combined)
```

```python
correlation_plot('trade_balance', 'hourly_rate')
```

There is a slight negative correlation between the trade balance and wages in this dataset. The shaded area is a 95% confidence interval 
of the regression. Since the area is so extreme, I would say there is little to no correlation between these traits.

```python
correlation_plot('trade_balance', 'working_hours')
```

Again there is a slight negative correlation between trade balance and working hours, but because of the wide area covered by the 95% 
confidence interval, I doubt there is any real correlation here either.
